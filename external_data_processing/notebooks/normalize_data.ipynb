```python
import os
import glob
import pandas as pd
import numpy as np
```

```python
path_to_mean = r'C:\Niloofar\SCR\posture-prediction-models-main\data\mean-std_20_4_2025'
```

```python
mean_data = pd.read_csv(os.path.join(path_to_mean, 'mean.csv'), header=None)
std_data = pd.read_csv(os.path.join(path_to_mean, 'std.csv'), header=None)
```

```python
mean_data
```

```python
Means = pd.concat(101*[mean_data], axis=0)
Means.reset_index(inplace=True, drop=True)

Stds = pd.concat(101*[std_data], axis=0)
Stds.reset_index(inplace=True, drop=True)
```

```python
f_means = [210.18638340655707, 289.51228353291765, 779.1597626742252]
f_stds = [317.8268718114922, 229.63004567188818, 564.7016492256631]
w_mean = 73.7
w_std = 9.83031026977277
h_mean = 177.8
h_std = 4.072468538859448
```

```python

os.mkdir(r"C:\Niloofar\SCR\posture-prediction-models-main\external_data_processing\\converted_data\\Final_data")

for i in range(1):
    load_path = r"C:\Niloofar\SCR\posture-prediction-models-main\external_data_processing\converted_data\s1_filtered"
    save_path = r"C:\Niloofar\SCR\posture-prediction-models-main\external_data_processing\\converted_data\\Final_data\\s1"
    os.mkdir(save_path)
    all_files = glob.glob(load_path + "\*.csv")

    # should not imposing the effect of the test_subject in calculation fo mu and sigma for training
    for filename in all_files:
        df = pd.read_csv(filename, header=None)
        # df = pd.read_csv(r'D:\\Master\\term2\\Social Robotics\\Project\\code\\s1_filtered\\s1_1_f.csv', header=None)
        # df.columns = list(range(42))
        df_normal = (df.iloc[:, :41]-Means.iloc[:, :41])/Stds.iloc[:, :41]
        df = pd.concat([df_normal, df.iloc[:, 41]], axis=1)
        df.iloc[5, -1] = (df.iloc[5, -1] - w_mean) / w_std  # weight normalization
        df.iloc[6, -1] = (df.iloc[6, -1] - h_mean) / h_std  # height normalization
        df.iloc[0, -1] = (df.iloc[0, -1] - f_means[0]) / f_stds[0]
        df.iloc[1, -1] = (df.iloc[1, -1] - f_means[1]) / f_stds[1]
        df.iloc[2, -1] = (df.iloc[2, -1] - f_means[2]) / f_stds[2]
        df.to_csv(save_path+'\\'+filename.split('\\')[-1].split('.')[0]+'n'+'.csv', sep=',', index=False, header=None)
```

```python
dfs = []
for i in range(1):
    path = r"C:\Niloofar\SCR\posture-prediction-models-main\external_data_processing\converted_data\s1_filtered"
    all_files = glob.glob(path + "\*.csv")
    
    for filename in all_files:
        df = pd.read_csv(filename, header=None)
        dfs.append(df)
        
combined_df = pd.concat(dfs, ignore_index=True)
combined_df
```
